```python
from lxml import etree
from selenium import webdriver
import json
from selenium.webdriver.support.select import Select
from scholarly import scholarly
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver import ActionChains
import time
import csv
import pandas as pd
from google_scholar_by_id import get_scholar_data
from google_scholar_by_id import parse
scholar_id_list = pd.read_csv('/Users/wenfei/Desktop/profiles_umich.csv',names=['url','name'])
scholar_id_dict = scholar_id_list.to_dict()
lis=[]
for i in range(0,1000): # set your target range. If we want all the data from umich, set the range(26288). Notice that this will takes a long time. 
    url = scholar_id_dict['url'][i]
    y = scholar_id_dict['name'][i] # name
    print(y)
    x = url.split('=')[-1]
    print(x)
    try:
        res = get_scholar_data(ids=x,name=y)
        print(res)
        lis.append(res)
    except:
        pass
with open("google_scholar_data_id.csv", mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(lis[0].keys())
    for res in lis:
        writer.writerow(res.values())
```
